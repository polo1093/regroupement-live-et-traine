# Train V3 Slim Mass Dataset

Run all cells to generate a V3-rich synthetic dataset, derive the current slim features from `shared_decision_features`, train the four stage models, and export a `v3_slim_latest` bundle.

The notebook calls the same pipeline as:

```bash
python launch.py train-slim --force
```

````python
from pathlib import Path
import json
import sys

ROOT = Path.cwd()
if not (ROOT / "Traine_aide_decission").exists():
    ROOT = ROOT.parent

TRAIN_DIR = ROOT / "Traine_aide_decission"
for path in (ROOT, TRAIN_DIR):
    if str(path) not in sys.path:
        sys.path.insert(0, str(path))

from experiments.v3_slim_mass_synthetic_pipeline import run_mass_synthetic_pipeline
from shared_decision_features import FEATURES_BY_STAGE

print(ROOT)
print(json.dumps({stage: list(features) for stage, features in FEATURES_BY_STAGE.items()}, indent=2))
````

````python
# Edit these values if needed, then run all cells.
OUTPUT_ROOT = ROOT / "outputs" / "readiness" / "v3_slim_mass_synthetic"
TOTAL_ROWS = 1_000_000
ROWS_PER_STAGE = None  # Use 1_000_000 here if you want 1M rows per stage instead of total.
CHUNK_SIZE = 100_000
RANDOM_SEED = 42
FORCE = True
TRAIN = True
EXPORT = True
````

````python
result = run_mass_synthetic_pipeline(
    output_root=OUTPUT_ROOT,
    total_rows=TOTAL_ROWS,
    rows_per_stage=ROWS_PER_STAGE,
    chunk_size=CHUNK_SIZE,
    random_seed=RANDOM_SEED,
    force=FORCE,
    train=TRAIN,
    export=EXPORT,
)

print(json.dumps({
    "status": result.get("status"),
    "rich_dir": result.get("rich_dir"),
    "slim_dir": result.get("slim_dir"),
    "model_root": result.get("model_root"),
    "export_dir": result.get("export_dir"),
    "generated_rows": result.get("generated", {}).get("total_rows"),
    "rows_per_stage": result.get("generated", {}).get("rows_per_stage"),
}, indent=2))
````
